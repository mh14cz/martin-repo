```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

def draw_base(ax):
    A = Circle((0, 0), 1.5, fill=False)
    B = Circle((1.2, 0), 1.5, fill=False)

    ax.add_patch(A)
    ax.add_patch(B)

    ax.text(-1.2, 1.6, "A", fontsize=12)
    ax.text(2.0, 1.6, "B", fontsize=12)

    ax.set_aspect("equal")
    ax.set_xlim(-3, 4)
    ax.set_ylim(-3, 3)
    ax.axis("off")

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# --- 1. Some A are B ---
draw_base(axs[0])

intersection = Circle((0.6, 0), 0.9, color="gray", alpha=0.5)
axs[0].add_patch(intersection)

axs[0].set_title("Some A are B")

# --- 2. Not all A are B ---
draw_base(axs[1])

a_not_b = Circle((0, 0), 1.5, color="gray", alpha=0.5)
axs[1].add_patch(a_not_b)

b_cover = Circle((1.2, 0), 1.5, color="white", zorder=3)
axs[1].add_patch(b_cover)

axs[1].set_title("Not all A are B")

# --- 3. Some A are not B ---
draw_base(axs[2])

a_not_b = Circle((0, 0), 1.5, color="gray", alpha=0.5)
axs[2].add_patch(a_not_b)

b_cover = Circle((1.2, 0), 1.5, color="white", zorder=3)
axs[2].add_patch(b_cover)

axs[2].set_title("Some A are not B")

plt.tight_layout()
plt.show()
```
