```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Ellipse
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

def draw_base(ax, title):
    # Base ellipses (outlines)
    ell_a = Ellipse(( -1,  0), width=5, height=2.8, fill=False)
    ell_b = Ellipse((  1,  0), width=5, height=2.8, fill=False)

    ax.add_patch(ell_a)
    ax.add_patch(ell_b)

    ax.text(-1, 1.5, "A", fontsize=12)
    ax.text( 1, 1.5, "B", fontsize=12)

    ax.set_title(title)
    ax.set_aspect("equal")
    ax.set_xlim(-4, 4)
    ax.set_ylim(-3, 2)
    ax.axis("off")

# --- Case 1: Some A are B ---
ax = axes[0]
draw_base(ax, "Some A are B")    
ax.text( 0, 0, "+1", fontsize=12)



# --- Case 2: Not all A are B ≡ Some A are not B  ---

ax = axes[1]
draw_base(ax, "Not all A are B ≡ Some A are not B")    
ax.text( -2.5, 0, "+1", fontsize=12)
ax.text(-3,-2.5, " “Not all A are B” and “Some A are not B” \nboth mean: “There is at least one A that is not B.”")



# 
# Make room at the bottom for figure-level text
fig.subplots_adjust(bottom=0.25)
fig.text(
    0.15, 0.25,
    "“Not all A are B” ≡ “Some A are not B”\n"
    "therefore\n"
    "“Some A are B” ≢ “Some A are not B”\n"
    "therefore\n"
    "“Some A are B” ≢ “Not all A are B”",
    ha="left", va="bottom", fontsize=10
)

plt.show()
```
