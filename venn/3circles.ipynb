```python
from matplotlib import pyplot as plt
from matplotlib.patches import Circle, Ellipse
from matplotlib_venn import venn2, venn3

```

```python
fig, ax = plt.subplots(figsize=(6, 6))

# Define circle parameters
radius = 1.5
centers = [
    (-1.0, 0.0),   # left
    ( 1.0, 0.0),   # right
    ( 0.0, 1.2)    # top
]

# Draw circles
for (x, y) in centers:
    circle = Circle((x, y), radius,
                    fill=False,
                    linewidth=2)
    ax.add_patch(circle)




ax.text(-2.8, 0, "Python developers", ha='right')
ax.text(2.8, 0, "JavaScript developers", ha='left')
ax.text(0, 2.9, "Logical thinkes", ha='center')


# Formatting
ax.set_aspect('equal')
ax.set_xlim(-4, 4)
ax.set_ylim(-3, 4)
ax.axis('off')

plt.show()
```
