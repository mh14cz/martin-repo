```python
from matplotlib import pyplot as plt
from matplotlib.patches import Circle, Ellipse
from matplotlib_venn import venn2, venn3
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

# Draw circles
# circles = [
#     Circle((-1,  1), 1.2, fill=False),
#     Circle(( 1,  1), 1.2, fill=False),
#     Circle((-1, -1), 1.2, fill=False),
#     Circle(( 1, -1), 1.2, fill=False),
# ]
#
# for c in circles:
#    ax.add_patch(c)
#

ells = [
    Ellipse((-1,  0.2), width=5, height=2.8, angle=-45),
    Ellipse(( 0,  1), width=5, height=2.8, angle=-45),
    Ellipse(( 0,  1), width=5, height=2.8, angle=45),
    Ellipse(( 1,  0.2), width=5, height=2.8, angle=45),

]    

for e in ells:
    
    e.set_fill(False)
    ax.add_patch(e)

# Labels for sets
ax.text(-2.8, 1.5, "Fire",  fontsize=12, weight="bold")
ax.text( -1.5,  2.8, "Earth", fontsize=12, weight="bold")
ax.text(0.95, 2.8, "Water", fontsize=12, weight="bold")
ax.text( 2.5, 1.5, "Air",   fontsize=12, weight="bold")
ax.text( 1.2, -0.5, "cave",   fontsize=10, weight="normal")
ax.text( -0.2, -1.5, "explosion",   fontsize=10, weight="normal")

items1 = "cave, cave explosion, cloud, explosion, mud, steam, volcano "
items2 = "underwater cave, underwater cave explosion, underwater explosion, underwater volcano"
ax.text( -3, -2.5, items1,   fontsize=10, weight="normal")
ax.text( -3, -2.8, items2,   fontsize=10, weight="normal")

ax.set_aspect("equal")
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.axis("off")

plt.show()
```
